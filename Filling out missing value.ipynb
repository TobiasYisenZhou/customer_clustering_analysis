## Feature engineering -- changing data value and type to wanted ones

```python
# finding everything ending with spend
all_columns=df20000.columns
all_columns=list(all_columns)
all_columns

all_spend=[ i for i in all_columns if re.search(r"spend$", i)]
all_spend
```

```python
# finding everything ending with trans（frequency)

all_trans=[ i for i in all_columns if re.search(r"trans$", i)]
all_trans
```

```python
# changing columns with spend as ending 

for i in all_spend:
    
# 1. changing ABCD... into numbers
    df20000[i]=df20000[i].replace({"A":10, "B":11,"C":12,"D":13,
                        "E":14,"F":15,"G":16,"H":17,"I":18,"J":19,"K":20,"L":21,"M":22,
                        "N":23,"O":24,"P":35,"Q":26,"R":27,"S":28,"T":29,"U":30,"V":31,
                         "W":32,"X":33,"Y":34,"Z":35})
    
# 2. changing column's data type into float
    dtype={i:float}
    df20000=df20000.astype(dtype)

```

## Creating two list, one for all numeircal columns, one for all categorical columns

```python
# Creating a list with name of all columns that contain numerical values

num_columns=["Number of Children (Enhanced)","Age in 2-year ranges - Person 1 (Enhanced)",
             "Target Home Market Value 2.0 Code","Advantage Length of Residence","Advantage Household Age (Enhanced)",
             "Advantage Target Income 3.0","Discretionary Spending Income","Target Net Worth 4.0 Code"]

print(len(all_columns))
```

```python
num_columns.extend(all_spend)
num_columns.extend(all_trans)
print(len(num_columns))
```

```python
# Creat a list with name of all columns with categorical values 
cat_columns=[ i for i in all_columns if i not in num_columns]
print(len(cat_columns))
```

```python
# Check. Expect "True"
len(all_columns)==len(num_columns)+len(cat_columns)
```

## Filling out missing value 

```python
# filling in "Missing" for all categorical columns
df1=df20000[cat_columns].fillna("0", axis=1)
df1.isnull().sum()
```

```python
# fill in mean for numerical columns

num_columns_average={}
for i in num_columns:
    inter=df20000[i].dropna()
    num_columns_average[i]=inter.mean()
 
num_columns_average
```

```python
num_fillna={}
for i in num_columns_average.keys():
    a=df20000.loc[:,i]
    a=a.fillna(num_columns_average[i])
    num_fillna[i]=a
    
#type(num_fillna["sc_home_furnishing_stores_22-24m_trans"].values)
df2=pd.DataFrame(num_fillna)
df2
```

```python
# cancatinating num tanble and cat table together
df_filled=pd.concat([df1,df2], axis=1)
df_filled.isnull().sum()
```
