```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import glob
```

```python
# loading all files and creating a united dataframe 
path = r'D:\CU\fall 2021\Practicum\data1' 
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.info()
```

```python
frame.head()
```

```python
# dropping unwanted columns
#frame.drop(labels=["Unnamed: 0"],axis=1, inplace=True)
frame.reset_index()
frame.head()
```

```python
# Getting a sense of how many values are missing
frame.isnull().sum()
```

1. 79% of the responses miss Number of Children. However, one cannot tell whether it is because the respondent does not have any, or due to missing data.
2. 

```python
# Some columns could use numerical value to represent. Changing A B C D into numbers
frame["Advantage Target Income 3.0"]=frame["Advantage Target Income 3.0"].replace({"A":10, "B":11,"C":12,"D":13})
frame["Discretionary Spending Income"]=frame["Discretionary Spending Income"].replace({"A":10, "B":11,"C":12})
frame["Target Net Worth 4.0 Code"]=frame["Target Net Worth 4.0 Code"].replace({"A":10, "B":11})
```

```python
# Changing each columns to the right datatype
frame.info()
```

```python
data_type={"Family composition (Enhanced)":str, "Gender code - Person 1":str, "Advantage Dwelling Type":str, "Advantage Home Owner":str, "Advantage Target Income 3.0": float,"Discretionary Spending Income":float,"Target Net Worth 4.0 Code":float}
frame=frame.astype(data_type)
frame.info()
```

```python
#frame.to_csv(r"D:\CU\fall 2021\Practicum\df1.csv")
```

```python
# taking out numerical values, and look at their correlation
num_columns=["Number of Children (Enhanced)","Age in 2-year ranges - Person 1 (Enhanced)",
             "Target Home Market Value 2.0 Code","Advantage Length of Residence","Advantage Household Age (Enhanced)",
             "Advantage Target Income 3.0","Discretionary Spending Income","Target Net Worth 4.0 Code"]
corrscore=frame[num_columns].corr()
sns.heatmap(corrscore) #seaborn
```
